```python
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
```

```python
test_df = pd.read_csv('test.csv')
```

```python
train_df['Gender'] = train_df['Sex']
```

```python
train_df = pd.get_dummies(train_df, columns=['Gender'])
```

```python
train_df.head()
```

```python
test_df['Gender'] = test_df['Sex']
```

```python
test_df = pd.get_dummies(test_df, columns=['Gender'])
```

```python
test_df.head()
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].mean())
```

```python
train_df.info()
```

```python
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].mean())
```

```python
test_df.info()
```

```python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Gender_female', 'Gender_male']
```

```python
import matplotlib.pylab as plt
```

```python
import seaborn as sns
```

```python
corr_data = train_df[['Pclass', 'Age', 'SibSp', 'Parch', 'Gender_female', 'Gender_male', 'Survived']].corr()
```

```python
sns.heatmap(corr_data)
```

```python
sns.pointplot(x='Pclass', y='Survived', hue='Sex', data=train_df)
```

```python
trainX = train_df[features]
```

```python
trainY = train_df['Survived']
```

```python
testX = test_df[features]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(trainX, trainY)
```

```python
prediction = model.predict(testX)
```

```python
prediction
```

```python
submission = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': prediction})
```

```python
submission.to_csv('submission.csv', index=False)
```
