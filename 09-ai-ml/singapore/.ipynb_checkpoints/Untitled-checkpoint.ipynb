```python
class Station:
    def __init__(self, name, line):
        self.name = name 
        self.line = line 
        self.connections = []

    def __str__(self):
        return self.name + " Line: " + self.line

class Edge:
    def __init__(self, dest, dur):
        self.dest = dest
        self.dur = dur
```

```python
stations = []
```

```python
for row in data.values:
    # print(row)
    # station = Station(row[2], 
    lines = row[3].split('/')
    for line in lines:
        station = Station(row[2], line)
        stations.append(station)
```

```python
for s in stations:
    print(s)
```
